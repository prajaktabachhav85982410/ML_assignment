```python
df.drop(['Surname', 'RowNumber','CustomerId'] , axis = 1 , inplace=True)
```

```python
# perform label encoding 
from sklearn.preprocessing import LabelEncoder
df['Geography'] = LabelEncoder().fit_transform(df['Geography'])
df['Gender'] = LabelEncoder().fit_transform(df['Gender'])
```

## split the data into x and y 

```python
x =df.drop(['Exited','HasCrCard'] , axis=1)
y = df['Exited']
```

```python
# split the data into itain test split 
from sklearn.model_selection import train_test_split

x_train , x_test, y_train , y_test = train_test_split(x , y, train_size=0.7 , random_state=12345)
```

# model training 

```python
def train_model_catboost():
    from catboost import CatBoostClassifier

    model = CatBoostClassifier()
    result = model.fit(x_train , y_train )
    return result
```

```python
def train_model_adaboost():
    from sklearn.ensemble import AdaBoostClassifier

    model = AdaBoostClassifier()
    result = model.fit(x_train , y_train )
    return result
```

```python
def train_model_xgb():
    from xgboost import XGBClassifier

    model = XGBClassifier()
    result = model.fit(x_train , y_train )
    return result
```

```python
models = [
    (train_model_catboost() , 'Catboost'), 
    (train_model_adaboost() , 'adaboost'),
    (train_model_xgb() , 'xgboost')
]
```

### evalute the model

```python
def evaluate_model(model , model_name):
    from sklearn.metrics import accuracy_score , precision_score ,recall_score , f1_score

    y_pred = model.predict(x_test)
    y_true = y_test
    accuracy = accuracy_score(y_true , y_pred)
    precision = precision_score(y_true , y_pred)
    recall = recall_score(y_true , y_pred)
    f1 = f1_score(y_true , y_pred)

    return model_name , accuracy , precision , recall , f1
    
```

```python
data = []
for (model , model_name) in models:
    data.append(evaluate_model(model, model_name))
result = pd.DataFrame(data , columns = ['Algorithm', 'Accuracy', 'Precision', 'Recall', 'F1'])
result 
```
