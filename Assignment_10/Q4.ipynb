#  split the data into x and y 
`

```python
x = df.drop(['K', 'Ca', 'Type'], axis=1)
y = df['Type']
```

# split the data into train test data 

```python
# split the data into train test data 
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x , y , train_size=0.7 , random_state=123456)
```

# model building 

```python
def train_model_catboost():
    from catboost import CatBoostClassifier

    model = CatBoostClassifier()
    result = model.fit(x_train , y_train )
    return result
```

```python
def train_model_adaboost():
    from sklearn.ensemble import AdaBoostClassifier

    model = AdaBoostClassifier()
    result = model.fit(x_train , y_train )
    return result
```

```python
# def train_model_xgb():
#     from xgboost import XGBClassifier

#     model = XGBClassifier()
#     result = model.fit(x_train , y_train )
#     return result
```

```python
models = [
    (train_model_catboost() , 'Catboost'), 
    (train_model_adaboost() , 'adaboost')
    # (train_model_xgb() , 'xgboost')
]
```

## evaluation of model

```python
def evaluate_model(model , model_name):
    from sklearn.metrics import accuracy_score , precision_score ,recall_score , f1_score

    y_pred = model.predict(x_test)
    y_true = y_test
    accuracy = accuracy_score(y_true , y_pred)
    precision = precision_score(y_true , y_pred,average='micro')
    recall = recall_score(y_true , y_pred,average='micro')
    f1 = f1_score(y_true , y_pred,average='micro')

    return model_name , accuracy , precision , recall , f1
    
```

```python
data = []
for (model , model_name) in models:
    data.append(evaluate_model(model, model_name))
result = pd.DataFrame(data , columns = ['Algorithm', 'Accuracy', 'Precision', 'Recall', 'F1'])
result 
```
