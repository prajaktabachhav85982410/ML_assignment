```python
df_cleaned = df1.dropna()
```

### split the data into x and y

```python
x = df_cleaned[['male', 'cigsPerDay' , 'BPMeds', 'prevalentStroke' , 'prevalentHyp' ,
'diabetes' ,'totChol' ,'sysBP']]

y =df_cleaned['TenYearCHD']
```

### split the data into train and test sets

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=12345)
```

### LogisticRegression Model training

```python
from sklearn.linear_model import LogisticRegressionCV

model_lg = LogisticRegressionCV(max_iter=1000)
model_lg.fit(x_train, y_train)
```

## naive bayes model training 

```python
from sklearn.naive_bayes import GaussianNB

model_nb = GaussianNB()

model_nb.fit(x_train, y_train)
```

### KNN Modeltraining

```python
from sklearn.neighbors import KNeighborsClassifier

model_knn = KNeighborsClassifier(n_neighbors=7)
model_knn.fit(x_train, y_train)
```

### SVM Model training

```python
from sklearn.svm import SVC

model_svc = SVC(C=8.0)
model_svc.fit(x_train, y_train)
```

### model evaluation

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

def evaluate_model(model, model_name):
    # get the predictions
    y_pred = model.predict(x_test)

    # ground truth
    y_true = y_test
    
    print(f"---- {model_name} evaluation ----")
    
    # get the confusion matrix
    cm = confusion_matrix(y_true, y_pred)
    print(f"confusion matrix: ")
    print(cm)

    # get the accuracy score
    print(f"accuracy : {accuracy_score(y_true, y_pred) * 100:.2f}%")

    # get the precision
    print(f"precision: {precision_score(y_true, y_pred):.2f}")

    # get the recall
    print(f"recall: {recall_score(y_true, y_pred):.2f}")

    # get the f1 score
    print(f"F1 Score: {f1_score(y_true, y_pred):.2f}")
```

```python
evaluate_model(model_lg, 'Logistic Regression')
```

```python
evaluate_model(model_nb, 'Naive Bayes')
```

```python
evaluate_model(model_knn, 'KNN')
```

```python
evaluate_model(model_svc, 'SVM')
```
